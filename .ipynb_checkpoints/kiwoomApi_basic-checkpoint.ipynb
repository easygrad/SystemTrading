```python
# 로그인 

from pykiwoom.kiwoom import *
# 엄청난 오류 발생!
# 1. 'QAxWidget' object has no attribute 'OnEventConnect' : 32bit 가상환경에서 실행해야함! anaconda prompt에서 새로운 가상환경 생성, 필요한 것들 재설치
# 2. No module named 'PyQt5.QAxContainer' / DLL load failed while importing QAxContainer : pyqt5 버전을 낮춰줘야함! pip install pyqt5==5.13
# https://corytips.tistory.com/206

kiwoom = Kiwoom()
kiwoom.CommConnect(block=True)
print("블로킹 로그인 완료")
```

```python
# 사용자 정보

account_num = kiwoom.GetLoginInfo("ACCOUNT_CNT")        # 전체 계좌수
accounts = kiwoom.GetLoginInfo("ACCNO")                 # 전체 계좌 리스트
user_id = kiwoom.GetLoginInfo("USER_ID")                # 사용자 ID
user_name = kiwoom.GetLoginInfo("USER_NAME")            # 사용자명
keyboard = kiwoom.GetLoginInfo("KEY_BSECGB")            # 키보드보안 해지여부
firewall = kiwoom.GetLoginInfo("FIREW_SECGB")           # 방화벽 설정 여부

print(account_num)
print(accounts)
print(user_id)
print(user_name)
print(keyboard)
print(firewall)
```

```python
# 종목코드

kospi = kiwoom.GetCodeListByMarket('0')
kosdaq = kiwoom.GetCodeListByMarket('10')
etf = kiwoom.GetCodeListByMarket('8')

print(len(kospi), "\n", kospi)
print("\n")
print(len(kosdaq), "\n", kosdaq)
print("\n")
print(len(etf), "\n", etf)
```

```python
# 테마

themeGroup = kiwoom.GetThemeGroupList(1)
print(themeGroup)
```

```python
# 종목명

name = kiwoom.GetMasterCodeName("005930")
print(name)
```

```python
# 연결상태

state = kiwoom.GetConnectState()
if state == 0:
    print("미연결")
elif state == 1:
    print("연결완료")
```

```python
# 상장주식수

stock_cnt = kiwoom.GetMasterListedStockCnt("005930")
print("삼성전자 상장주식수: ", stock_cnt)

# 감리구분

kamri = kiwoom.GetMasterConstruction("005930")
print(kamri)

# 상장일

ipoDate = kiwoom.GetMasterListedStockDate("005930")
print(ipoDate)
print(type(ipoDate))

# 종목상태

stockState = kiwoom.GetMasterStockState("005930")
print(stockState)
```

```python
# 주식계좌
accounts = kiwoom.GetLoginInfo("ACCNO")
stock_account = accounts[0]

# 삼성전자, 10주, 시장가주문 매수
kiwoom.SendOrder("시장가매수", "0101", stock_account, 1, "005930", 10, 0, "03", "")

# 삼성전자, 10주, 시장가주문 매도
kiwoom.SendOrder("시장가매도", "0101", stock_account, 2, "005930", 10, 0, "03", "")
```

```python
# 싱글데이터 TR 사용하기

df = kiwoom.block_request("opt10001",
                          종목코드="005930",
                          output="주식기본정보",
                          next=0)
print(df)
```

```python
# 멀티데이터 TR 사용하기

df = kiwoom.block_request("opt10081",
                          종목코드="005930",
                          기준일자="20200424",
                          수정주가구분=1,
                          output="주식일봉차트조회",
                          next=0)
print(df.head())
df.append(df)
```
