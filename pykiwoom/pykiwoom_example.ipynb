```python
# 비 실시간 전략 구현
# https://wikidocs.net/book/1173

# 전종목 일봉 데이터 다운로드
from pykiwoom.kiwoom import *
import datetime
import time

```

```python
# 로그인
kiwoom = Kiwoom()
kiwoom.CommConnect()
```

```python
# 전종목 종목코드
kospi = kiwoom.GetCodeListByMarket('0')
kosdaq = kiwoom.GetCodeListByMarket('10')
codes = kospi + kosdaq
codes
```

```python
# 문자열로 오늘 날짜 얻기
now = datetime.datetime.now()
today = now.strftime("%Y%m%d")
print(now)
print(today)
```

```python
# 전종목의 일봉 데이터 엑셀파일로 저장하기
for i, code in enumerate(codes):
    print(f"{i}/{len(codes)} {code}")
    df = kiwoom.block_request("opt10081",
                             종목코드=code,
                             기준일자=today,
                             수정주가구분=1,
                             output="주식일봉차트조회",
                             next=0)
    
    out_name = f"{code}.xlsx"
    df.to_excel(r'C:\Users\allma\Desktop\sysTrading_allma\kiwoomApi_exercise\{}'.format(out_name))
    time.sleep(0.6) # 0.2초로 했더니 제한 메시지 뜸. 너무 짧아서 그런듯
    
# 짚고 넘어갈 문법 f-string : 문자열 입력시 {} 안에 직접 변수 이름을 입력
# .format() : 문자열 중간에 {} 안에 들어가라 변수 이름을 입력
```

```python
# 전 종목 일봉 데이터 머지(merge)
import pandas as pd
import os

path = r"C:\Users\allma\Desktop\sysTrading_allma\kiwoomApi_exercise"
flist = os.listdir(path)
xlsx_list = [x for x in flist if x.endswith('.xlsx')]
close_data = []

for xls in xlsx_list:
    code = xls.split('.')[0]
    df = pd.read_excel(r'C:\Users\allma\Desktop\sysTrading_allma\kiwoomApi_exercise\{}'.format(xls), engine='openpyxl') # XLRDError: Excel xlsx file; not supported? 안정성 이슈로 xlrd에서 xlsx 확장자 읽기 기능 지원 중단.. engine 설정해서 해결
    df2 = df[['일자', '현재가']].copy() # 얕은 복사
    df2.rename(columns={'현재가': code}, inplace=True) # inplace 디폴트 값은 False, 이 경우 저장되지 않음. 저장하려면 True
    df2 = df2.set_index('일자')
    df2 = df2[::-1] # 굳이 해줄 필요가 있나?
    close_data.append(df2)
    
# concat
df = pd.concat(close_data, axis=1)
df.to_excel('merge.xlsx')
```

```python
# 모멘텀 전략 종목 선정 : 최근에 가장 많이 오른 종목을 매수
import pandas as pd

df = pd.read_excel("merge.xlsx", engine='openpyxl')
df['일자'] = pd.to_datetime(df['일자'], format="%Y%m%d") # datetime 타입으로 변환, 판다스 편리한 기능 활용하기 위해
df = df.set_index('일자')

# 3개월 가격 모멘텀, 거래일 기준 60일

return_df = df.pct_change(60)
return_df.tail() # 끝에 5개 행만 출력
```

```python
# 데이터 프레임으로 만들기
s = return_df.loc["2020-06-22"]
momentum_df = pd.DataFrame(s)
momentum_df.columns = ["모멘텀"]
momentum_df.head(n=10)
```

```python
# 순위 매기기: 모멘텀 전략은 수익률 높은 종목 선정하는 방식
momentum_df['순위'] = momentum_df['모멘텀'].rank(ascending=False)
momentum_df.head(n=10)
```

```python
# 순위를 기준으로 정렬
momentum_df = momentum_df.sort_values(by='순위')
momentum_df[:30]
```

```python
# 종목명 추가하기
import pandas as pd
import time

df = pd.read_excel("momentum_list.xlsx", engine='openpyxl')
df.columns = ["종목코드", "모멘텀", "순위"]

# 종목코드 6자리가 숫자로 인식되면서 해제됨 / 처음에 해볼땐 숫자로 인식됐는데.. 그 다음 할때는 제대로 문자로 인식됨
codes = df["종목코드"]
names = [kiwoom.GetMasterCodeName(code) for code in codes]
df['종목명'] = pd.Series(data=names)
```

```python
# 매수하기
accounts = kiwoom.GetLoginInfo('ACCNO')
account = accounts[0]

for code in codes:
    ret = kiwoom.SendOrder("시장가매수", "0101", account, 1, code, 100, 0, "03", "")
    time.sleep(0.2)
    print(code, "종목 시장가 주문 완료")
```

```python
# KOA - 개발가이드 - 주문과 잔고처리
# LONG SendOrder(BSTR sRQName: 사용자구분명, BSTR sScreenNo: 화면번호, BSTR sAccNo: 계좌번호, Long nOrederType: 주문유형, BSTR sCode: 종목코드, LONG nQty: 주문수량, LONG nPrice: 주문가격 시장가 주문시 0 / 취소 주문시 0, BSTR sHogaGb: 거래구분 혹은 호가구분, BSTR sOrgOrderNo: 원주문번호 신규주문에는 공백이고 정정/취소시 입력)
# 주문유형 1 신규매수 / 2 신규매도 / 3 매수취소 / 4 매도취소 / 5 매수정정 / 6 매도정정

# 매수취소
for code in codes:
    ret = kiwoom.SendOrder("시장가매수취소", "0101", account, 3, code, 100, 0, "03", "")
    time.sleep(0.2)
    print(code, "종목 시장가 주문 완료")
```
